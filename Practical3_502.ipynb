```python
#HeatMap
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data=np.random.randint (1,100,size=(10,10))
print("The data to be plotted: \n")
print(data)
```

```python
#Plotting Heatmap
hm=sns.heatmap(data=data)
plt.show()
```

```python
hm = sns.heatmap(data=data,
                 vmin='30',
                 vmax='70')
plt.show()
```
